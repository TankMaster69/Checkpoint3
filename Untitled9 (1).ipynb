```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
```

```python
plot_correlation_map(df)
```

This function helps us to find the pairwise of each column in the df except for categorical values. The values the function calculate are regrouped in a subplot with a palette using seaborn libraby for a better result.In the right, we have an indicator with a color gradient. The more red the color is, the more the values will be correlated and inversly with blue."
